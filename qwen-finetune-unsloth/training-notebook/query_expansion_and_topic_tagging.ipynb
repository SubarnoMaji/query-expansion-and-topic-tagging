```python
import json
from datasets import Dataset

TOPIC_HIERARCHY = {
    "Politics": ["India", "UK", "USA", "China", "Russia", "Global"],
    "Sports": ["Cricket", "Football", "Basketball", "Tennis", "Olympics"],
    "Technology": ["Artificial Intelligence","Machine Learning","Software Development","Cybersecurity","Blockchain"],
    "Business": ["Startups","Finance","Stock Market","Economy","E-commerce"],
    "Entertainment": ["Movies","TV Shows", "Music", "Celebrities", "OTT Platforms"],
    "Science": ["Physics", "Biology", "Space", "Climate", "Research"],
    "Health": ["Fitness", "Nutrition", "Mental Health", "Diseases", "Medicine"],
    "Education": ["Exams","Universities","Online Courses","Careers","Research"],
    "General": ["Chitchat", "Greetings", "Meta", "Clarification", "Other"]
}

actor_prompt_instructions = f"""
expanded_query Rules

You must rewrite the userâ€™s latest query into a fully explicit, standalone question by resolving:
- Pronouns (he, she, his, her, they)
- Ellipsis (â€œwhat aboutâ€¦â€, â€œand UK?â€, â€œsame for himâ€)
- Implicit references to earlier entities, countries, roles, or topics

Guidelines:
- Preserve the userâ€™s last original intent
- Inject missing entities from prior dialogue only if clearly implied
- Do NOT hallucinate new entities
- If the query is already self-contained, return it unchanged
- If the user intent is unclear, return the best minimal expansion
- Preserve the original grammatical person and sentence form
- If the user asks a question, the expansion must remain a question
- If the user uses imperative or fragment form, preserve it

--------------------------------
Topic Classification Rules

You MUST assign a topic using ONLY the following hierarchy.
- Select exactly ONE level_1
- Select exactly ONE level_2 under that level_1
- Do NOT invent new topics
- If no specific domain fits, use:
  level_1 = "General"
  level_2 = "Other"

Allowed Topics:

{json.dumps(TOPIC_HIERARCHY, indent=2)}

--------------------------------
Output Format (STRICT JSON ONLY):

{{
  "messages": [
    {{"role": "user", "content": "..."}},
    {{"role": "assistant", "content": "..."}},
    {{"role": "user", "content": "..."}}
  ],
  "labels": {{
    "expanded_query": "...",
    "topic": {{
      "level_1": "<must be one key from the hierarchy>",
      "level_2": "<must be a valid subtopic of level_1>"
    }}
  }}
}}

Do not add explanations or extra text.
""".strip()

alpaca_prompt = """Below is an instruction for a task.

### Instruction:
{INSTRUCTION}

### Input:
{INPUT}

### Response:
{OUTPUT}
"""

EOS_TOKEN = tokenizer.eos_token

def extract_instruction_input_output(entries):
    inputs, outputs = [], []

    for entry in entries:
        messages = entry.get("messages", [])
        labels = entry.get("labels", {})
        expanded_query = labels.get("expanded_query", "").strip()

        if not expanded_query:
            continue

        # Format dialogue
        dialogue = ""
        for m in messages:
            role = m.get("role", "").capitalize()
            content = m.get("content", "")
            dialogue += f"{role}: {content}\n"

        topic = labels.get("topic", {})
        output_json = json.dumps(
            {
                "expanded_query": expanded_query,
                "topic": {
                    "level_1": topic.get("level_1", ""),
                    "level_2": topic.get("level_2", "")
                }
            },
            ensure_ascii=False,
            indent=2
        )

        inputs.append(dialogue.strip())
        outputs.append(output_json)

    return {"input": inputs, "output": outputs}

# -----------------------------
# Formatting Function (NO KeyError)
# -----------------------------
def formatting_prompts_func(examples):
    texts = []
    for inp, out in zip(examples["input"], examples["output"]):
        prompt = alpaca_prompt.format_map({
            "INSTRUCTION": actor_prompt_instructions,
            "INPUT": inp,
            "OUTPUT": out
        })
        texts.append(prompt + EOS_TOKEN)
    return {"text": texts}
```

```python
def split_prompt_and_response(text):
    marker = "### Response:"
    if marker not in text:
        return text.strip(), ""

    prompt, response = text.split(marker, 1)
    return prompt.strip(), response.strip()

def strip_response(example):
    prompt, _ = split_prompt_and_response(example["text"])
    return {"text": prompt}

datasets["val-nores"] = datasets["val"].map(strip_response)
```

```python
def run_inference(prompts, max_new_tokens=256):
    inputs = tokenizer(
        prompts,
        return_tensors="pt",
        padding=True,
        truncation=True
    ).to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        use_cache=True
    )

    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    return decoded
```

```python
import json
from tqdm import tqdm

results = []

batch_size = 8

val_prompts = datasets["val-nores"]["text"]
val_full = datasets["val"]["text"]

for i in tqdm(range(0, len(val_prompts), batch_size)):
    batch_prompts = val_prompts[i:i + batch_size]
    batch_full = val_full[i:i + batch_size]

    model_outputs = run_inference(batch_prompts, max_new_tokens=256)

    for prompt, full_text, model_out in zip(batch_prompts, batch_full, model_outputs):
        _, ground_truth = split_prompt_and_response(full_text)

        results.append({
            "prompt": prompt,
            "ground_truth": ground_truth,
            "model_output": model_out
        })
```

```python
with open("val_inference_results.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(f"Saved {len(results)} validation results to val_inference_results.json")
```
